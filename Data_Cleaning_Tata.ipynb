```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("/content/drive/MyDrive/tata/Online Retail Data Set.xlsx")
df
```

```python
df.info()
```

```python
df['UnitPrice']=df['UnitPrice'].astype(str).str.replace('-', '').astype(float)
df['UnitPrice']
```

```python
df['Revenue']=df['Quantity']*df['UnitPrice']
```

```python
def fill_revenue(row):
    if row['Quantity'] <= 0 or row['UnitPrice'] <= 0:
        return None
    else:
        return row['Quantity'] * row['UnitPrice']


df['Revenue'] = df.apply(fill_revenue, axis=1)
```

```python
df['Revenue']
```

```python
def fill_revenue(row):
    if row['Quantity'] <= 0 or row['UnitPrice'] <= 0:
        return -1 * row['Quantity'] * row['UnitPrice']
    else:
        return None


df['Revenue Refund'] = df.apply(fill_revenue, axis=1)
```

```python
df.info()
```

```python
df = df[df['Description'].notna()]
```
